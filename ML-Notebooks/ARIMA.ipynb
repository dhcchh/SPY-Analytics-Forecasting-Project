```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.stattools import adfuller
```

```python
df_full = pd.read_csv('../Data/spy_full.csv')
df_month = pd.read_csv('../Data/spy_month.csv')
df_week = pd.read_csv('../Data/spy_week.csv')
df_full['date'] = pd.to_datetime(df_full['date'])
df_month['date'] = pd.to_datetime(df_month['date'])
df_week['date'] = pd.to_datetime(df_week['date'])
df_week_test = df_full[~df_full['date'].isin(df_week['date'])]
df_month_test = df_full[~df_full['date'].isin(df_month['date'])]
```

```python
plt.plot(df_full['date'], df_full['close_price']);
```

```python
adf = adfuller(df_full['close_price'])
print(f'ADF statistic: {adf[0]:.4f}\n'
      f'p-value: {adf[1]:.4f}\n')
```

```python
plt.plot(df_full['date'], df_full['close_price'].diff());
```

```python
adf = adfuller(df_full['close_price'].diff().dropna())
print(f'ADF statistic: {adf[0]:.4f}\n'
      f'p-value: {adf[1]:.4f}\n')
```

```python
plt.plot(df_full['date'], np.log(df_full['close_price']).diff());
```

```python
plt.plot(df_full['date'],df_full['log_return']);
```

```python
adf = adfuller(df_full['log_return'].dropna())
print(f'ADF statistic: {adf[0]:.4f}\n'
      f'p-value: {adf[1]:.4f}\n')
```
